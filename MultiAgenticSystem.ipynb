```python
from transformers import pipeline

llm = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=200
)
```

```python
from typing import TypedDict

class UIState(TypedDict):
    brand_input: str
    theme: str
    accessibility: str
    component_code: str
    optimized_code: str
    documentation: str
```

```python
def generate(prompt):
    response = llm(prompt)
    return response[0]["generated_text"]
```

```python
def brand_agent(state: UIState):

    prompt = f"""
    Analyze this company branding.

    Branding:
    {state['brand_input']}

    Extract:
    - design style
    - colors
    - typography
    """

    result = generate(prompt)

    state["theme"] = result

    return state
```

```python
def theme_agent(state: UIState):

    prompt = f"""
    Generate Tailwind theme config.

    Theme:
    {state['theme']}
    """

    result = generate(prompt)

    state["theme"] = result

    return state
```

```python
def accessibility_agent(state: UIState):

    prompt = f"""
    Improve accessibility for this UI theme.

    Theme:
    {state['theme']}
    """

    result = generate(prompt)

    state["accessibility"] = result

    return state
```

```python
def component_agent(state: UIState):

    prompt = f"""
    Generate React TypeScript button component.within Codes in React Component

    Requirements:
    - TailwindCSS
    - shadcn/ui style
    - loading state
    - variants

    Theme:
    {state['theme']}
    """

    result = generate(prompt)

    state["component_code"] = result

    return state
```

```python
def optimization_agent(state: UIState):

    prompt = f"""
    Optimize this React component.

    Component:
    {state['component_code']}
    """

    result = generate(prompt)

    state["optimized_code"] = result

    return state
```

```python
def docs_agent(state: UIState):

    prompt = f"""
    Create markdown documentation.

    Component:
    {state['optimized_code']}
    """

    result = generate(prompt)

    state["documentation"] = result

    return state
```

```python
from langgraph.graph import StateGraph, END

workflow = StateGraph(UIState)

workflow.add_node("brand", brand_agent)
workflow.add_node("theme", theme_agent)
workflow.add_node("accessibility", accessibility_agent)
workflow.add_node("component", component_agent)
workflow.add_node("optimization", optimization_agent)
workflow.add_node("docs", docs_agent)

workflow.set_entry_point("brand")

workflow.add_edge("brand", "theme")
workflow.add_edge("theme", "accessibility")
workflow.add_edge("accessibility", "component")
workflow.add_edge("component", "optimization")
workflow.add_edge("optimization", "docs")
workflow.add_edge("docs", END)

app = workflow.compile()
```

```python
result = app.invoke({
    "brand_input": """
    Modern AI company.
    Dark futuristic UI.
    Purple and blue gradients.
    Minimal enterprise design.
    """
})
```
